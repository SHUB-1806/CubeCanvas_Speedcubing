## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
maxm, minm, sum = 0, 100, 0
marks = []
for x in range(10):
    n = int(input(f"Enter the marks of student {x+1}: "))
    while n<0 or n>100:
        n = int(input("Marks should be from 0 to 100: "))
    marks.append(n)
    maxm = max(maxm, n)
    minm = min(minm, n)
    sum += n

avg = sum/10
aboveAvg = 0
grades = {
    "A": 0,
    "B": 0,
    "C": 0,
    "D": 0
}
for i in marks:
    if i > avg:
        aboveAvg += 1
    
    if i>=90:
        grades["A"]+=1
    elif i>=75:
        grades["B"]+=1
    elif i>=60:
        grades["C"]+=1
    else:
        grades["D"]+=1

print("Max Marks: ", maxm)
print("Min Marks: ", minm)
print("Avg Marks: ", avg)
print("Students with above avg Marks: ", aboveAvg)
for a, b in grades.items():
    print(a, ": ", b)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]

validColors = {
    "R":0,
    "O":0,
    "G":0,
    "B":0,
    "W":0,
    "Y":0
}

solvedRow = []

i=0
for row in face:
    for color in row:
        if color not in validColors:
            print("(Invalid)", end=" ")
        else:
            print("X" if color=="R" else color, end=" ")
            validColors[color]+=1
    print("")
    i+=1
    if(len(set(row))==1):
        solvedRow.append(i)

highestFreq = max(validColors, key=validColors.get)

if len(solvedRow)==0:
    print("No row has all identical colors.")
else:
    for x in solvedRow:
        print("Row ", x, " has all identical colors.")

print("Most frequent color is: ", highestFreq)
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
vowels = ('a', 'e', 'i', 'o', 'u')
chars = {}
sentence = input("Enter the sentence: ")
wordsList = sentence.split()
numWords, numChar, numVowels = len(wordsList), 0, 0

for word in wordsList:
    numChar += len(word)
    for ch in word:
        if ch.lower() in vowels:
            numVowels+=1

        if ch.lower() in chars:
            chars[ch.lower()]+=1
        else:
            chars[ch.lower()]=1

noSpaces = "".join(wordsList)
noSpacesRev = noSpaces[::-1]
mostFreq = max(chars, key=chars.get)


if noSpaces.lower() == noSpacesRev.lower():
    print("Sentence is a palindrome.")
else:
    print("Sentence is not a palindrome.")

print("Number of words is: ", numWords)
print("Number of characters is: ", numChar)
print("Number of vowels is: ", numVowels)
print("Most frequent character is: ", mostFreq)
````
