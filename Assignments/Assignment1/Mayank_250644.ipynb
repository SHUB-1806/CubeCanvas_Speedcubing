## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
def input_marks():
    a = []
    for i in range(10):
        j = int(input(""))
        a.append(j)
    return a

def average(a):
    sum = 0
    for i in a:
        sum += i
    return sum/len(a)

def highest(a):
    maxi = 0
    for i in a:
        maxi = max(maxi, i)
    return maxi

def lowest(a):
    lowe = float('inf')
    for i in a:
        lowe = min(lowe, i)
    return lowe

def no_above_avg(a):
    count = 0
    avg = average(a)
    for i in a:
        if i > avg:
            count+=1
    return count    

def grades(a):
    b = []
    for i in a:
        if i >= 90:
            b.append("A")
        elif i >= 75:
            b.append("B")
        elif i >= 60:
            b.append("C")
        else:
            b.append("D")
    return b

def gradedisplay(a):
    b = grades(a)
    c = {
        'A': 0,
        'B': 0,
        'C': 0,
        'D': 0
    }
    for i in b:
        c[i]+=1
    print (c)

a = input_marks()
print("Highest Marks:", highest(a))
print("Lowest Marks:", lowest(a))
print("Averag Marks:", average(a))
print("No. of students scoring above avg are:", no_above_avg(a))
gradedisplay(a)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
face = [
 ["R","G","G"],
 ["W","R","G"],
 ["B","B","Y"]
]

def printit():
    for i in face:
        for j in i:
            print(j, end="")
        print("")

def freq():
    freq = {
        'R' : 0,
        'G' : 0,
        'B' : 0,
        'Y' : 0,
        'W' : 0,
        'O' : 0,
    }
    for i in face:
        for j in i:
            freq[j]+=1
    return freq

def r_to_x():
    for i in range(3):
        for j in range(3):
            if face[i][j] == 'R':
                face[i][j] = 'X'

def common():
    j = freq()
    return max(j, key=j.get)
    
def chek():
    for i in face:
        count = 0
        a = i[0]
        for j in i:
            if a == j:
                count+=1
        if count == 3:
            return "Yes, a row contains identical colors"
    return "No, a row doesn't contain identical colors"

print(printit())
print("Freq:", freq())
print("The most common color is:", common())
r_to_x()
print(chek())
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
a = str(input())
def count_chars():
    count = 0
    for i in a:
        count+=1
    return count

def count_words():
    count = 0
    token = a.split()
    for i in token:
        count+=1
    return count
    

def count_vowels():
    count = 0
    for i in a:
        if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
            count+=1
    return count

def freq():
    freq = {}
    for i in a:
        if i in freq:
            freq[i]+=1
        else:
            freq[i] = 1
    return freq

def most_freq():
    mf = ""
    occ = 0
    fre = freq()
    for i in a:
        if occ <= fre[i]:
            occ = fre[i]
            mf = i
    return mf

def pali():
    a.lower().replace(" ", "")
    if a == a[::1]:
        return True
    return False

print("No. of chars:",count_chars())
print("No. of words:",count_words())
print("No. of vowels:",count_vowels())
print("Freq of chars:",freq())
print("Most freq char:",most_freq())
print("If it is Palindrome:",pali())
````
