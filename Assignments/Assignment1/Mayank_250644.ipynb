## Question 1: Student Performance Analyzer (Easy)

Write a program that:

1. Takes marks of 10 students as input.
2. Stores them in a list.
3. Calculates:
   - Highest marks
   - Lowest marks
   - Average marks
   - Number of students scoring above average
4. Assign grades:
   - A : >= 90
   - B : >= 75
   - C : >= 60
   - D : otherwise
5. Display the grade distribution using a dictionary.


````python
# Write your solution here
def input_marks():
    a = []
    for i in range(10):
        j = int(input(""))
        a.append(j)
    return a

def average(a):
    sum = 0
    for i in a:
        sum += i
    return sum/len(a)

def highest(a):
    maxi = 0
    for i in a:
        maxi = max(maxi, i)
    return maxi

def lowest(a):
    lowe = float('inf')
    for i in a:
        lowe = min(lowe, i)
    return lowe

def no_above_avg(a):
    count = 0
    avg = average(a)
    for i in a:
        if i > avg:
            count+=1
    return count    

def grades(a):
    b = []
    for i in a:
        if i >= 90:
            b.append("A")
        elif i >= 75:
            b.append("B")
        elif i >= 60:
            b.append("C")
        else:
            b.append("D")
    return b

def gradedisplay(a):
    b = grades(a)
    c = {
        'A': 0,
        'B': 0,
        'C': 0,
        'D': 0
    }
    for i in b:
        c[i]+=1
    print (c)

a = input_marks()
print("Highest Marks:", highest(a))
print("Lowest Marks:", lowest(a))
print("Averag Marks:", average(a))
print("No. of students scoring above avg are:", no_above_avg(a))
gradedisplay(a)
````

## Question 2: Color Grid Processor (Medium)

A Rubik's Cube face is represented as:

```python
face = [
 ["R","R","G"],
 ["W","R","G"],
 ["B","B","Y"]
]
```

Write a program that:

1. Prints the grid neatly.
2. Counts the frequency of every color.
3. Replaces every occurrence of `R -> X`.
4. Determines the most common color.
5. Checks whether any row contains all identical colors.


````python
# Write your solution here
face = [
 ["R","G","G"],
 ["W","R","G"],
 ["B","B","Y"]
]

def printit():
    for i in face:
        for j in i:
            print(j, end="")
        print("")

def freq():
    freq = {
        'R' : 0,
        'G' : 0,
        'B' : 0,
        'Y' : 0,
        'W' : 0,
        'O' : 0,
    }
    for i in face:
        for j in i:
            freq[j]+=1
    return freq

def r_to_x():
    for i in range(3):
        for j in range(3):
            if face[i][j] == 'R':
                face[i][j] = 'X'

def common():
    j = freq()
    return max(j, key=j.get)
    
def chek():
    for i in face:
        count = 0
        a = i[0]
        for j in i:
            if a == j:
                count+=1
        if count == 3:
            return "Yes, a row contains identical colors"
    return "No, a row doesn't contain identical colors"

print(printit())
print("Freq:", freq())
print("The most common color is:", common())
r_to_x()
print(chek())
````

## Question 3: Text Analytics Tool (Medium)

Given a sentence entered by the user:

1. Count total characters.
2. Count total words.
3. Count vowels.
4. Count frequency of every character using a dictionary.
5. Find the most frequently occurring character.
6. Check whether the sentence is a palindrome after removing spaces and converting to lowercase.

Example:

```text
Input:
Never Odd Or Even

Output:
Palindrome : True
```


````python
# Write your solution here
a = str(input())
def count_chars():
    count = 0
    for i in a:
        count+=1
    return count

def count_words():
    count = 0
    token = a.split()
    for i in token:
        count+=1
    return count
    

def count_vowels():
    count = 0
    for i in a:
        if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
            count+=1
    return count

def freq():
    freq = {}
    for i in a:
        if i in freq:
            freq[i]+=1
        else:
            freq[i] = 1
    return freq

def most_freq():
    mf = ""
    occ = 0
    fre = freq()
    for i in a:
        if occ <= fre[i]:
            occ = fre[i]
            mf = i
    return mf

def pali():
    a.lower().replace(" ", "")
    if a == a[::1]:
        return True
    return False

print("No. of chars:",count_chars())
print("No. of words:",count_words())
print("No. of vowels:",count_vowels())
print("Freq of chars:",freq())
print("Most freq char:",most_freq())
print("If it is Palindrome:",pali())
````

````python
# Mini Project Solution
class Member:
    def __init__(self, name, roll, branch, yr, avg):
        self.name = name
        self.roll_no = roll
        self.branch = branch
        self.yr = yr
        self.avg = avg

    def display(self):
        print("Name:", self.name)
        print("Roll No.:", self.roll_no)
        print("Branch:", self.branch)
        print("Year:", self.yr)
        print("Average Time:", self.avg)

    def update_avg(self, new_avg):
        self.avg = new_avg

class Event:
    def __init__(self, name, date, venue):
        self.name = name
        self.date = date
        self.venue = venue
        self.list = []

    def register_member(self, member):
        self.list.append(member)

    def remove_member(self, member):
        self.list.remove(member)
    
    def display_participants(self):
        for i in self.list:
            print(i)

class Club:
    def __init__(self, name):
        self.name = name
        self.list = []
        self.events = []

    def add_mem(self, member):
        self.list.append(member)

    def rem_mem(self, member):
        self.list.remove(member)
    
    def search(self, r_no):
        for i in self.list:
            if i.roll_no == r_no:
                print(i.name)
                return
        print("Not found")
    
    def create_event(self, event):
        self.events.append(event)
        
    def count_branch(self):
        count = {}
        for i in self.list:
            if i.branch in count:
                count[i.branch] += 1
            else:
                count[i.branch] = 1
        print(count)

    def display_eve(self):
        for i in self.events:
            print(i.name, end = " ")
        print()

    def display_mem(self):
        for i in self.list:
            print(i.name, end = " ")
        print()
    
    def top_3(self):
        top_members = sorted(self.list, key=lambda x: x.avg, reverse=True)[:3]

        for member in top_members:
            print(member.name, member.avg)

    def all_avg(self):
        sum = 0
        for i in self.list:
            sum += i.avg
        print(sum/len(self.list))
    
    def search_by_branch(self, branch):
        list = []
        for i in self.list:
            if i.branch == branch:
                list.append(i.name)
        print(list)
    
    def ranking(self):
        top_members = sorted(self.list, key=lambda x: x.avg, reverse=True)

        for member in top_members:
            print(member.name, member.avg)

club = Club("Chess Club")

a = Member("A", 120609, "CE", 2, 15.9)
b = Member("B", 120608, "ECE", 2, 25.6)
c = Member("C", 120605, "EE", 2, 14.9)
d = Member("D", 120602, "CSE", 2, 17.3)

eventt = Event("Master Cubers", "12-6-26", "MPR")

club.add_mem(a)
club.add_mem(b)
club.add_mem(c)
club.add_mem(d)

club.create_event(eventt)

for i in club.list:
    eventt.register_member(i)

club.all_avg()
club.search(120609)
club.display_eve()
club.display_mem()
club.top_3()
club.count_branch()
club.search_by_branch("ECE")
club.ranking()
````
