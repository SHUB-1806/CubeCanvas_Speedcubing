# Question 4: Interpolation Investigation (Hard)

Different interpolation techniques produce different results.


## Part A (Theory)

Research and explain:

1. Nearest Neighbor Interpolation
2. Bilinear Interpolation
3. Bicubic Interpolation
4. Lanczos Interpolation

For each method discuss:
- Working principle
- Advantages
- Disadvantages
- Suitable applications


### Theory Answer Space

1. Nearest Neighbor interpolation
    * `Working principle`
    Nearest Neighbor interpolation assigns the value of the nearest pixel to the new pixel's place.
    For every new pixel, we get the nearest original pixel and copy its color directly.
    * `Advantages`
    Very fast to compute, easy to implement and requires the least processing power.

    * `Disadvantages`
        Gives a blocky and pixelated image, forming jagged edges and poor image quality when enlarged.

    * `Suitable Application`
        It can be used in pixel art and retro style designs or for real time graphics were speed matters or for quick image previews.
2. Bilinear interpolation
    * `Working Principle`
      Bilinear Interpolation gets the new pixel's value by using the four nearest neighboring pixels by interpolating them horizontally and vertically and then using the weighted average.

    * `Advantages`
      It produces smoother images with less pixelation than nearest neighbor interpolation but still remaining relatively fast.
    * `Disadvantage`
      It still has blurry details and sharp edges become softer.
    * `Suitable Applications`
      It can be used for general image resizing or computer graphics and video processing.
3. Bicubic Interpolation
    * `Working Principle`
      Bicubic Interpolation uses the 16 nearest pixels (4x4 neighborhood surrounding target pixel) to calculate value of new pixel, But instead of using linear averaging, it fits cubic curves through neighboring pixels (i.e. cubic polynomial functions).

    * `Advantages`
      It gives smoother images but with better edge preservation and has a more natural appearance and higher image quality than Bilinear.
    * `Disadvantages`
      It is slower than Bilinear as it requires more computation.
    * `Suitable Applications`
      In digital photography, image editing software and professional image processing.
4. Lanczos Interpolation
    * `Working Principle`
      Lanczos Interpolation uses a sinc-based mathematical function to calculate the new pixel's value. Instead of considering just nearby pixels, it uses a large neighborhood around the target pixel.

    * `Advantages`
      It gives the best image quality with preserved fine details with better edge and texture handling.
    * `Disadvantages`
      It requires extremely high number of computations and hence is the slowest method of them all.
    * `Suitable Applications`
      It can be used in high quality image enlargement, printing and publishing and for astronomical and scientific imaging.

## Part B (Coding)

Using the same image:

1. Resize the image to 100x100.
2. Resize it back to 800x800 using:
   - Nearest Neighbor
   - Bilinear
   - Bicubic
   - Lanczos
3. Save all outputs.


````python
# Write your solution here

from PIL import Image

img = Image.open('images.jpg')
resized_100 = img.resize(
    (100, 100),
    Image.Resampling.LANCZOS
)

nearest = resized_100.resize(
    (800, 800),
    Image.Resampling.NEAREST
)

bilinear = resized_100.resize(
    (800, 800),
    Image.Resampling.BILINEAR
)

bicubic = resized_100.resize(
    (800, 800),
    Image.Resampling.BICUBIC
)

lanczos = resized_100.resize(
    (800, 800),
    Image.Resampling.LANCZOS
)

nearest.save("nearest.jpg")
bilinear.save("bilinear.jpg")
bicubic.save("bicubic.jpg")
lanczos.save("lanczos.jpg")
````
